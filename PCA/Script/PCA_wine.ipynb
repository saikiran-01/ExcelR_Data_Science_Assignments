There are some quite correlation between variables. For example the correlation between flavanoids and dilution is pretty high (78%). Thus we can remove that variable from our dataset. However this method is long and tedious. Hence we use PCA method for Dimensionality Reduction.

Dimensionality Reduction with PCA

```python
# normalizing the data
wine_norm = StandardScaler().fit_transform(wine)
```

```python
pca = PCA(n_components=13)
```

```python
principalComponents = pca.fit_transform(wine_norm)
```

```python
PC = range(1, pca.n_components_+1)
plt.bar(PC, pca.explained_variance_ratio_, color='blue')
plt.xlabel('Principal Components')
plt.ylabel('Variance %')
plt.xticks(PC)
```

```python
PCA_components = pd.DataFrame(principalComponents)
```

```python
plt.scatter(PCA_components[0], PCA_components[1], alpha=.3, color='blue')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.show()
```

As shown in the bar graph, the most of varianve is put in the first 2 components. Since there is not much variance present from 3rd component, lets just the first 2 componets in our analysis. The scatter plot given an indication that there may be 3 clusters present

Finding out the optimal number of clusters

```python
wcss = []
```

```python
for i in range(1, 15):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(PCA_components.iloc[:,:3])
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 15), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

The scree plot levels off at k=3 and let's use it to determine the clusters

K-means clustering

```python
model = KMeans(n_clusters=3)
model.fit(PCA_components.iloc[:,:2])
```

```python
labels = model.predict(PCA_components.iloc[:,:2])
```

```python
plt.scatter(PCA_components[0], PCA_components[1], c=labels)
plt.show()
```
