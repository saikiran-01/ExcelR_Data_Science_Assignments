As the data is not in transaction formation We are using transaction Encoder

```python
df=pd.get_dummies(book)
df.head()
```

Apriori Algorithm

1) Association rules with 10% Support and 70% confidence

```python
# With 10% Support
frequent_itemsets = apriori(df, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
# with 70% confidence
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.7)
rules
```

```python
#Sorting Values By Decending Order.
rules.sort_values('lift',ascending = False)[0:20]
```

```python
# Lift Ratio > 1 is a good influential rule in selecting the associated transactions
rules[rules.lift>1]
```

```python
# visualization of obtained rule
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```

2) Association rules with 15% Support and 85% confidence

```python
# With 15% Support
frequent_itemsets = apriori(df, min_support=0.15, use_colnames=True)
frequent_itemsets
```

```python
# with 85% confidence
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.85)
rules
```

```python
rules.sort_values('lift',ascending = False)[0:20]
```

```python
# visualization of obtained rule
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```

3) Association rules for Support 5% & confidence 90%

```python
# With 5% Support
frequent_itemsets = apriori(df, min_support=0.05, use_colnames=True)
frequent_itemsets
```

```python
# with 90% confidence
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.9)
rules
```

```python
rules.sort_values('lift',ascending = False)[0:20]
```
